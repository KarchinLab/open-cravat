# Open-CRAVAT Quickstart

Open-CRAVAT is a modular, local, and open-source variant interpretation platform.

Open-CRAVAT is composed of the core programs and modules. The core programs provide the operating environment for the modules and the modules do the operations of format conversion, gene mapping, annotation, filtration, and reporting.

The minimum set of Open-CRAVAT modules for running Open-CRAVAT in your computer is called "base modules" and they are composed of the modules of input format converters, hg38 gene mapper, analysis aggregator, Excel and text file reporters, analysis result web viewer, and web interface for analysis, Open-CRAVAT Store, and system settings.

Let's install the core programs and the base modules with the following two commands. The hg38 gene mapper in the base modules is about 8GB big, so it can take some time.

Open-CRAVAT requires Python3.5 or higher.

```python
!pip3 install open-cravat
```

```python
!cravat-admin install-base
```

Now Open-CRAVAT is ready! At this point, it can do gene mapping and sequence ontology annotation with the hg38 mapper.

Let's do a test analysis. We need an input file. Open-CRAVAT provides a test input file for your convenience. Go to a folder where you want to store CRAVAT analysis files and run the following.

```python
!cravat-admin make-example-input . 
```

This will create `example_input` file in the current directory.

Now, let's run an Open-CRAVAT analysis on this input file with the following command.

```python
!cravat example_input
```

This will create a bunch of files. But, the final output is example_input.xlsx. Open it and see the analysis result.

Open-CRAVAT provides a web-based analysis result viewer. The above cravat run will have created the file `example_input.sqlite` which is a SQLite3 database file of the analysis result. Run the below command to see it in the web result viewer.

```python
!cravat-view example_input.sqlite
```

Now, let's add ClinVar annotation for this same input.

First, get the ClinVar module for Open-CRAVAT with the following command.

```python
!cravat-admin install clinvar
```

By the way, Open-CRAVAT has many analysis modules available. To see all the currently available Open-CRAVAT modules, run the below command.

```python
!cravat-admin ls -a
```

Let's run ClinVar annotation for the same input.

```python
!cravat example_input
```

Note that we did not specify to do ClinVar annotation. This is because Open-CRAVAT by default runs all annotation modules. To see which annotation modules are currently installed in your system, run the below command.

```python
!cravat-admin ls
```

To specify the annotation modules to run, you can give `-a` option to `cravat`. For example, to run ClinVar annotation only, you can run the below command.

```python
!cravat example_input -a clinvar
```

Note that `-a` option came after the input file path and that `clinvar` was used instead of ClinVar. `clinvar` is the registered name of the ClinVar module in Open-CRAVAT and this is what we used when we installed the module and also what we see in the output of `cravat-admin ls` (with or without `-a` option).
